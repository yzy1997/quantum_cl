```python
import qiskit

qc = qiskit.QuantumCircuit(3)
qc.x(0)

qc.draw('mpl')
```

```python
qc = qiskit.QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.x(2)
qc.draw('mpl')
```

```python
from qiskit.circuit.library import TwoLocal
from math import pi

reference_circuit = TwoLocal(3, 'ry', 'cz', reps=3, entanglement='linear', insert_barriers=True)
theta_list = [pi/12, pi/6, pi/4, pi/3, 5*pi/12, pi/2, 7*pi/12, 2*pi/3, 3*pi/4, 5*pi/6, 11*pi/12, pi]
reference_circuit = reference_circuit.assign_parameters(theta_list)

reference_circuit.decompose().draw('mpl')
```

```python
from qiskit.circuit.library import ZZFeatureMap

data = [0.1, 0.2]

zz_feature_map_reference = ZZFeatureMap(feature_dimension=2, reps=2, entanglement='linear')
zz_encoded_data = zz_feature_map_reference.assign_parameters(data)
zz_encoded_data.decompose().draw(output='mpl')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_vector
from IPython.display import display  # 确保在 Jupyter Notebook 显示多个图像

# 设定旋转角度
theta_values = np.linspace(0, np.pi, 5)  # 选择 5 个不同的 θ 进行可视化

# 遍历不同的旋转角度
for theta in theta_values:
    # 创建量子电路并施加 Ry(θ) 旋转
    qc = QuantumCircuit(1)
    qc.ry(theta, 0)

    # 获取旋转后的量子态
    state = Statevector.from_instruction(qc)

    # 计算 Bloch 球上的 (x, y, z) 坐标
    bloch_x = 2 * np.real(state[0] * np.conj(state[1]))  # X = 2 Re(αβ*)
    bloch_y = 2 * np.imag(state[0] * np.conj(state[1]))  # Y = 2 Im(αβ*)
    bloch_z = np.abs(state[0])**2 - np.abs(state[1])**2  # Z = |α|^2 - |β|^2

    # 绘制 Bloch 球向量并显示
    display(plot_bloch_vector([bloch_x, bloch_y, bloch_z], title=f'Bloch Vector for R_y({theta:.2f})'))

plt.show()  # 确保 Jupyter Notebook 能显示所有图像
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_vector
from IPython.display import display  #  Jupyter Notebook 

# 
theta_values = np.linspace(0, np.pi, 5)  #  5   

# 
for theta in theta_values:
    #  Ry() 
    qc = QuantumCircuit(1)
    qc.ry(theta, 0)

    # 
    state = Statevector.from_instruction(qc)

    #  Bloch  (x, y, z) 
    bloch_x = 2 * np.real(state[0] * np.conj(state[1]))  # X = 2 Re(*)
    bloch_y = 2 * np.imag(state[0] * np.conj(state[1]))  # Y = 2 Im(*)
    bloch_z = np.abs(state[0])**2 - np.abs(state[1])**2  # Z = ||^2 - ||^2

    #  Bloch 
    display(plot_bloch_vector([bloch_x, bloch_y, bloch_z], title=f'Bloch Vector for R_y({theta:.2f})'))

plt.show()  #  Jupyter Notebook 
```
